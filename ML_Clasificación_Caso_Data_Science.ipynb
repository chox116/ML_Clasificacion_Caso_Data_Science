Definimos unas funciones que nos ayudaran a obtener informaciÃ³n de las metricas y mejorarlas

```python
def tupla_clase_prediccion(y_real, y_pred): #creamos una funciÃ³n  que le daremos dos listas, el valor verdadero y la etiqueta de predicciÃ³n
    return list(zip(y_real, y_pred))        #nos arrojarÃ¡ una lista de parejas(producto cartesiano)

def VP(clases_reales, predicciones):                                #definirmos el VP (verdadero positivo)
    par_clase_prediccion = tupla_clase_prediccion(clases_reales, predicciones) #creamos una tupla con los c
    return len([obs for obs in par_clase_prediccion if obs[0]==1 and obs[1]==1])  #internamente crea una lista donde va a meter en ella los que tenga (1,1)

def VN(clases_reales, predicciones):                               #definimos VN (verdadero negativo)
    par_clase_prediccion = tupla_clase_prediccion(clases_reales, predicciones)
    return len([obs for obs in par_clase_prediccion if obs[0]==0 and obs[1]==0])#internamente crea una lista donde va a meter en ella los que tenga (0,0)

def FP(clases_reales, predicciones):                              #definimos FP(Falso positivo)
    par_clase_prediccion = tupla_clase_prediccion(clases_reales, predicciones)
    return len([obs for obs in par_clase_prediccion if obs[0]==0 and obs[1]==1])#internamente crea una lista donde va a meter en ella los que tenga (0,1)

def FN(clases_reales, predicciones):                             #definimos FN(Falso negativo) este FN es de interes pues queremos disminuir este para evitar gente muerta
    par_clase_prediccion = tupla_clase_prediccion(clases_reales, predicciones) #se utiliza la sencibiliad(recall) pues si obtenemos Sencibilidad=1=VP/(VP+FP) entonces FP=0
    return len([obs for obs in par_clase_prediccion if obs[0]==1 and obs[1]==0])#internamente crea una lista donde va a meter en ella los que tenga (1,0)
```

```python
tupla_clase_prediccion(y_test_clas, predicciones_rl)[:20] #nos arroja el objetivo de prueba en pareja junto con su predicciÃ³n
```

```python
print("""
Verdaderos Positivos: {}
Verdaderos Negativos: {}
Falsos Positivos: {}
Falsos Negativos: {}
""".format(
    VP(y_test_clas, predicciones_rl),
    VN(y_test_clas, predicciones_rl),
    FP(y_test_clas, predicciones_rl),
    FN(y_test_clas, predicciones_rl)
))
```

### Evaluando las metricas

```python
from sklearn import metrics

metrics.accuracy_score(y_test_clas, predicciones_rl)
```

```python
def precision(clases_reales, predicciones):   #esta funciÃ³n no  estÃ¡ definida en sklearn por lo que se define en el modelo
    vp = VP(clases_reales, predicciones)
    fp = FP(clases_reales, predicciones)
    return vp / (vp+fp)

precision(y_test_clas, predicciones_rl)
```

```python
metrics.recall_score(y_test_clas, predicciones_rl) #sensibilidad
```

```python
metrics.f1_score(y_test_clas, predicciones_rl) #F1
```

```python
def proba_a_etiqueta(predicciones_proba,umbral=0.5):     #definimos una funcion que se guia por las probabilidades y el umbral para cambiar el resultado de nuestra predicciÃ³n
    predicciones=np.zeros([len(predicciones_proba), ]) #a partir de nuestro umbral, no es lo mismo umbral=0.5 a umbral=0.7( a 0.7 si prob>0.7 entonces sera la etiqueta 1)
    predicciones[predicciones_proba[:,1]>=umbral]=1    #En umbral definimos el punto que tomara para la probabilidad de que sea 1 o 0, normalmente es 0.5
    return predicciones

proba_a_etiqueta(predicciones_proba)
```

```python
def evaluar_umbral(umbral):                      #esta funcion genera los datos de presicion, sensibilidad y f1 respecto a un umbral
    predicciones_en_umbral=proba_a_etiqueta(predicciones_proba, umbral) #se genera una funcion sencibilidad(umbral), presicion(umbral) F1(umbral)
    precision_umbral=precision(y_test_clas, predicciones_en_umbral)
    sensibilidad_umbral=metrics.recall_score(y_test_clas, predicciones_en_umbral)
    F1_umbral=metrics.f1_score(y_test_clas, predicciones_en_umbral)
    return (umbral,precision_umbral, sensibilidad_umbral, F1_umbral)

umbrales = np.linspace(0., 1., 1000)
evaluaciones=pd.DataFrame([evaluar_umbral(x) for x in umbrales],
                          columns = ["umbral","precision","sensibilidad","F1"])
evaluaciones.head(20)
```

Como lo que buscamos es que haya ganancias, entonces vamos a considerar la sensibilidad pues nos interesa mas que haya falsos negativos(que diga que es perdida pero es ganancia) que falsos positivos(ganancia pero realmente es perdida). Por lo que seleccionaremos el umbral donde sea maximo la sencibilidad pero cuando el umbral es cercano a cero y nos da un valor de sensibilidad de 1 podria tratarse de un sobre ajuste, por lo que se utilizara de criterio el maximo de F1

### Optimizando hiperparametros con busqueda de malla

```python
from sklearn.model_selection import GridSearchCV #llamamos busqueda por malla(GRID) y busqueda aleatoria(Random)
from sklearn.model_selection import cross_validate

resultados={} #creamos un diccionario
```

```python
estimador_reglog=LogisticRegression()
print(estimador_reglog.__doc__) #pedimos que imprima informaci+on del estimador
estimador_reglog.get_params() #pedimos que umprima los parametros que usa el estimador
```

```python
parametros_busqueda_reglog={ #damos algunos parametros para la busqueda por malla
    "penalty":['l2','None'],
    "random_state":[0,1,5,20,50],
    "solver": ["liblinear", "newton-cholesky",'saga']
}
```

```python
reglog_grid=GridSearchCV(estimator=estimador_reglog, #gridsearch es busqueda por malla, el estimador no tiene los parametros
                    param_grid=parametros_busqueda_reglog, #en param_grid le pasamos la malla que es el diccionario param_busque_knn
                    scoring="f1", n_jobs=-1) #n_jobs es para ver que nucleos usar
```

```python
import time
from siuba import *
from siuba.dply.vector import *

def evaluar_modelo(estimador, X, y):
    resultados_estimador = cross_validate(estimador, X, y,
                     scoring="f1_micro", n_jobs=-1, cv=5) #pedimos micro pues los datos estan balanceados(distribuidos equitativamente)
    return resultados_estimador #en binaria se puede usar ROC AU, se usan 5 particiones de validacioon cruzada

def ver_resultados():             #promedia los resultados de la funcion anterior y nos da una tabla
    resultados_df=pd.DataFrame(resultados).T
    resultados_cols=resultados_df.columns
    for col in resultados_df:
        resultados_df[col]=resultados_df[col].apply(np.mean)
        resultados_df[col+"_idx"]=resultados_df[col]/resultados_df[col].max()
    return resultados_df>>arrange(-_.test_score,_.fit_time)
```

```python
start_time = time.time() #medimos el tiempo
 
reglog_grid.fit(X_train_clas,y_train_clas) #entrenamos el modelo con la busqueda de malla

end_time = time.time() #medimos el tiempo de uevo
elapsed_time = end_time - start_time #vemos cuanto tiempo nos tomo 

print(f"Elapsed time: {elapsed_time} seconds")

print(reglog_grid.best_score_) #pedimos el mejor score del modelo
print(reglog_grid.best_estimator_.get_params()) #pedimos que nos de los mejores parametros para el mejor score

resultados["reglog_gridsearch"] = evaluar_modelo(reglog_grid.best_estimator_, #entrenamos el modelo con los mejores hiperparametros
                                             X_train_clas,y_train_clas)

ver_resultados()
```
